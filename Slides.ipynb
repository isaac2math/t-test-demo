### * t distribution is the distribution of the sample mean, whose sample is drawn randomly from a normal - distributed population.
### * standard t distribution is uniquely defined by its degree of freedom (one degree of freedom is mapped to only one t distribution)
### * t distribution is heavy-tail distribution, whose tail is slightly heavier than normal distribution. (As degree of freedom increasing, t distribution will be infinitely similar to normal distribution, see the following graph)
	
### refer to the discussion of "heavy tail and outliers discussion" in tut 2

```python
## Setup      
df        = 30   # Degrees of freedom
alpha     = 0.05 # Significence level
top_yaxis = 0.4
```

## Two-tail t test

### $$ H_0 : \mu = \mu_0$$
### $$ H_1 : \mu \neq \mu_0$$
### $$ t = \frac{ \bar{x} - \mu_0}{se} \sim t_{n-1}$$

### Rejection region method : rejection of $H_0$ (t-stat on the left)

```python
tt = t_test_demo(df, alpha, top_yaxis, 'Two', 'True-Left')
tt.t_plot()
```

### Rejection region method : rejection of $H_0$ (t-stat on the right)

```python
tt = t_test_demo(df, alpha, top_yaxis, 'Two', 'True-Right')
tt.t_plot()
```

### Rejection region method : non-rejection of $H_0$

```python
tt = t_test_demo(df, alpha, top_yaxis, 'Two', 'False')
tt.t_plot()
```

### p-value method : compute two-tail p-value

```python
tt.t_pvalue_plot()
```
